##**‚ù§ PYTHON PROGRAM** üëá



```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# reading the file
data = pd.read_csv("project1.csv")

X = data[['Hours']]
y = data['Score']

model = LinearRegression()

# Training the model
model.fit(X, y)

# model's prediction
y_predict = model.predict(X)

# evaluation
mae = mean_absolute_error(y, y_predict)
mse = mean_squared_error(y, y_predict)
rmse = np.sqrt(mse)

def data_show():
  print(
        "\n1-- Showing Entire Data of Dataset \n"
        "2--Showing X - Actual Inputs \n"
        "3--Showing y - Actual Outputs \n"
        "4-- Models Prediction (predicted y - Output) \n"
        "5-- Evaluation of model \n"
        "6-- Testing for new data \n"
        "7-- showing the Actual graph \n"
        "8-- showing graph of models prediction \n"
        "9-- showing graph of Testing Data \n"
        "10-- Exit\n"
    )


def dimagh_walla_function():
  while True:
    global hours, new_predicted_score
    data_show()
    choice = input("\nWhat you want to do with above Dataset? ")

    if choice == "1":
        print(f'\nThe Entire Data of Dataset:\n{data}')

    elif choice == "2":
        print(f'\nThe Actual inputs:\n{X}')

    elif choice == "3":
        print(f'\nThe Actual Outputs:\n{y}')

    elif choice == "4":
        print(f'\nThe Predicted Output by a Model:\n{y_predict}')

    elif choice == "5":
        print("\n----Model Evaluation----")
        print("Mean Absolute Error (MAE): ", mae)
        print("Mean Squared Error (MSE): ", mse)
        print("Root Mean Squared Error (RMSE): ", rmse)

    elif choice == "6":
        ask_testing = input("\nDo you really want to (Test) Check the Model: ").lower()
        while ask_testing == "yes":
            hours = int(input("Please Enter Hours: "))
            new_predicted_score = model.predict([[hours]])[0]
            print(f"\nBased on {hours} Predicted Score: ", int(new_predicted_score))
            ask_testing = input("Do you really want to (Test) Check the Model again: ").lower()

    elif choice == "7":
        print("\n----Showing Actual Graph----")
        plt.plot(X, y, marker='o')
        plt.xlabel("Hours Studied")
        plt.ylabel("Score Scored")
        plt.title("Actual Data")
        plt.grid(True)
        plt.show()


    elif choice == "8":
        print("\n----Showing Graph of Models Prediction----")
        plt.plot(X, y_predict)
        plt.xlabel("Hours Studied")
        plt.ylabel("Score Scored")
        plt.title("Model Prediction")
        plt.grid(True)
        plt.show()


    elif choice == "9":
        print("\n----Showing Graph of Testing Data----")
        if hours is not None:
            plt.scatter(hours, new_predicted_score)
            plt.xlabel("New Hours")
            plt.ylabel("Predicted Score")
            plt.title("Testing Data")
            plt.grid(True)
            plt.show()

        else:
            print("Please test the model first (Option 6).")

    elif choice == "10":
        print("Thank You! you are exiting the program....")
        break

    else:
        print("Invalid Input")

dimagh_walla_function()
```
